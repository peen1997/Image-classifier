```python
import numpy as np
import matplotlib.pyplot as plt 
import keras 
from scipy import misc, ndimage
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator
%matplotlib inline 
```

```python
# plots images with labels within jupyter notebook
def plots(ims, figsize=(12,6), rows=1, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=10)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
```

```python
gen = ImageDataGenerator(rotation_range =10, width_shift_range= 0 , height_shift_range=0, shear_range=0.1,zoom_range = 0.3,channel_shift_range=10)
```

```python
imgpath = 'gallery collection/toilet problems/toilet41.jpg'
```

```python
#Obtain image 
image = np.expand_dims(plt.imread(imgpath),0)
plt.imshow(image[0])
```

```python
#Generate batches of auguamented images from this image 
aug_iter = gen.flow(image)
```

```python
#Get 10 samples 
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(20)]
```

```python
#plot images 
plots(aug_images,figsize=(20,7),rows=2)
```
